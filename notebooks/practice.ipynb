```python
frame = pd.DataFrame({
    'b': [4, 7, -3, 2],
    'a': [0, 1, 0, 1]
}, index = ['Row3', 'Row2', 'Row1', 'Row4'])

frame
```

```python
scores = pd.DataFrame({
    'Student': ['Alice', 'Bob', 'Charlie', 'David'],
    'Math': [85, 45, 92, 60],
    'Science': [70, 55, 88, 72]
})

scores

scores.sort_values(by = ['Math'], ascending=False, inplace=True)
scores

scores['Math_Rank'] = scores['Math'].rank(ascending=False)
scores
```

```python
def grade(score):
    return 'Pass' if score >= 60 else 'Fail'

scores['Status'] = scores['Math'].apply(grade)
scores
```

```python
scores.info()
```
